```python
from ortools.linear_solver import pywraplp

solver = pywraplp.Solver.CreateSolver('SCIP')

L = solver.IntVar(0, solver.infinity(), 'L')
P = solver.IntVar(0, solver.infinity(), 'P')
T = solver.IntVar(0, solver.infinity(), 'T')

solver.Add(2*L + P + T <= 100)
solver.Add(L + P + T <= 80)
solver.Add(L + P + 2*T <= 120)

solver.Add(L >= 10)
solver.Add(P >= 10)
solver.Add(T >= 10)

solver.Maximize(500*L + 300*P + 400*T)

status = solver.Solve()

if status == pywraplp.Solver.OPTIMAL:
    print("Laptops =", int(L.solution_value()))
    print("Phones =", int(P.solution_value()))
    print("Tablets =", int(T.solution_value()))
    print("Maximum Profit =", int(solver.Objective().Value()))
```
